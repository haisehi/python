```python
from matplotlib import pyplot as plt
```

```python
x = [5,2,9,4,7]
```

```python
y = [10,5,8,4,2]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.hist(y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Example")
plt.xlabel("Time (hr)")
plt.ylabel("Position (Km)")
plt.show()
```
