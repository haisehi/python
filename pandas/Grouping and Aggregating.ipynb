```python
#Splitting the data into groups based on some criteria.
#normalize dùng để chia theo phần trăm
df['SocialMedia'].value_counts(normalize=True)
```

```python
df['Country'].value_counts()
```

```python
# nhóm tất cả kết quả theo quốc gia
country_grp = df.groupby(['Country'])
```

```python
# lấy một nhóm theo một quốc gia cụ thể
country_grp.get_group('China')
```

```python
# lọc ra những mạng xã hội được dùng tại một quốc gia nào đó
country_grp['SocialMedia'].value_counts().loc['Viet Nam']
```

```python
# lọc ra lương trung bình ở Đức
country_grp['ConvertedComp'].median().loc['Germany']
```

```python
# dùng hàm tổng hợp
#country_grp['ConvertedComp'].agg(['median','mean'])
# theo một quốc gia cụ thể
country_grp['ConvertedComp'].agg(['median','mean']).loc['Viet Nam']
```

```python
# lọc ra số người dùng ngôn ngữ python để làm việc
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python')
```

```python
# tính tổng số người dùng ngôn ngữ python để làm việc
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```
