```python
#Python Pandas Tutorial (Part 6): Add/Remove Rows and Columns From DataFrames
```

```python
people = {
    "first":["Corey",'Jane','John'],
    "last":["Schafer",'Doe','Doe'],
    "email":["CoreyMSchafer@gmail.com",'JaneDoe@email.com','JohnDoe@email.com'],
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['full name'] = df['first']+' '+df['last'] #combining columns
df
```

```python
# removing columns
df.drop(columns = ['first','last'],inplace=True)
df
```

```python
#string split
#expand : chia giá trị được cắt thành 2 cột
df['full name'].str.split(' ', expand=True)
```

```python
# multiple columns
df[['first','last']] =  df['full name'].str.split(' ', expand=True)
df
```

```python
# add single row and ignore index
df._append({'first':'Tony'}, ignore_index= True)
```

```python
#create new data frame
people = {
    "first":["Tony",'Steve'],
    "last":["Starl",'Rogers'],
    "email":["IronMan@gmail.com","Cap@gmail.com"],
}
df2 = pd.DataFrame(people)
```

```python
df2
```

```python
#append data Frame
#sort columns
df._append(df2, ignore_index=True)
```

```python
df.drop(index=df[df['first']=='Steve'].index)
```
