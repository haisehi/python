```python
# Import các thư viện cần thiết
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Tạo dữ liệu mẫu
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Số lượng quảng cáo
y = 2 * X + np.random.randn(100, 1) * 2  # Doanh số bán hàng, có thêm noise
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Xây dựng mô hình Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Dự đoán doanh số bán hàng trên tập kiểm tra
y_pred = model.predict(X_test)
```

```python
# Trực quan hóa kết quả
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Số lượng quảng cáo')
plt.ylabel('Doanh số bán hàng')
plt.title('Mô hình Linear Regression')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Tạo dữ liệu mẫu
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Số lượng quảng cáo
y = 3 * X + np.random.randn(100, 1) * 2  # Doanh số bán hàng (có nhiễu)
```

```python
# Xây dựng mô hình Decision Tree
model = DecisionTreeRegressor()
model.fit(X, y)
```

```python
# Dự đoán doanh số bán hàng cho dữ liệu mới
new_data = np.array([[5]])  # Số lượng quảng cáo mới
predicted_sales = model.predict(new_data)
print("Dự đoán doanh số bán hàng cho số lượng quảng cáo mới:", predicted_sales[0])
```

```python
# Trực quan hóa mô hình
X_grid = np.arange(min(X), max(X), 0.01).reshape(-1, 1)
plt.scatter(X, y, color='blue')
plt.plot(X_grid, model.predict(X_grid), color='red')
plt.title('Decision Tree Regression')
plt.xlabel('Số lượng quảng cáo')
plt.ylabel('Doanh số bán hàng')
plt.show()
```

```python
# Tạo dữ liệu mẫu
np.random.seed(0)
data = pd.DataFrame({
    'Quang_cao': np.random.randint(50, 400, 500),  # Số lượng quảng cáo
    'Gia': np.random.randint(50, 300, 500),  # Giá cả
    'Mua_hang': np.random.randint(500, 3500, 500)  # Doanh số bán hàng
})
```

```python
data
```

```python
# Phân chia dữ liệu thành features (X) và target (y)
X = data[['Quang_cao', 'Gia']]
y = data['Mua_hang']
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
# Xây dựng mô hình Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Đánh giá mô hình trên tập kiểm tra
score = model.score(X_test, y_test)
print("Độ chính xác trên tập kiểm tra:", score)
```

```python
# áp dụng một mô hình hồi quy phi tuyến tính (Non-linear Regression) 
# bằng cách sử dụng Polynomial Regression để dự đoán doanh số bán hàng của một siêu thị mini với 1000 dòng dữ liệu.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Tạo dữ liệu mẫu
np.random.seed(0)
data = pd.DataFrame({
    'Quang_cao': np.random.randint(50, 200, 2000),  # Số lượng quảng cáo
    'Gia': np.random.randint(70, 400, 2000),  # Giá cả
    'Mua_hang': np.random.randint(600, 1500, 2000)  # Doanh số bán hàng
})
```

```python
# Phân chia dữ liệu thành features (X) và target (y)
X = data[['Quang_cao', 'Gia']]
y = data['Mua_hang']
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Áp dụng Polynomial Regression
poly = PolynomialFeatures(degree=2)  # Đa thức bậc 2
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
# Xây dựng mô hình Linear Regression
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
# Dự đoán doanh số bán hàng trên tập kiểm tra
y_pred = model.predict(X_test_poly)
```

```python
# Đánh giá mô hình
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Trực quan hóa mô hình
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Thực tế')
plt.ylabel('Dự đoán')
plt.title('Polynomial Regression')
plt.show()
```
